```python
string = input()
hashtb = {}
iteration = []

for i in string:

    if i != ' ' and i != '\n':

        if i not in hashtb:
            iteration.append(i)
            hashtb[i] = 1

        else:
            hashtb[i] += 1
maxelem = 0

iteration.sort()

for i in hashtb:
    maxelem = max(maxelem, hashtb[i])


while maxelem != 0:

    for i in iteration:

        if hashtb[i] == maxelem:  
            print('#', end = '')
            hashtb[i] -= 1
            tru_max = hashtb[i]

        else:
            print(' ', end = '')
    print()

    maxelem = tru_max 
print(''.join(iteration))
```

```python
# num 3
n = int(input())
diego = [int(i) for i in input().split()]
k = int(input())
colections = [int(i) for i in input().split()]

def ans(n,diego,k, colections):
    diego = list(set(diego))
    diego.sort()
    sort_colections = sorted(colections)

    hashmp = {}

    R,L = 0, 0
    count = 0

    for R in range(len(sort_colections)):
        while L < len(diego) and sort_colections[R] > diego[L]:
            count += 1
            L += 1
        hashmp[sort_colections[R]] = count

    for i in colections:
        if i in hashmp:
            print(hashmp[i])
        else:
            print(0)

ans(n,diego,k, colections)
        
```

```python
# number4
N = int(input())
k = int(input())
row = int(input())
left_or_right = int(input())

def ans():
    
    if left_or_right == 2:
        place = row * 2 
    else:
        place = row * 2 - 1
        
    
    
    if place + k <= N and place - k > 0:
        
        if (place + k) % 2 == 0:
            ans = [(place + k) // 2, 2] 
        else:
            ans = [(place + k) // 2 + 1, 1]
            
        
        if (place - k) % 2 == 0:
            q_ans = [(place - k) // 2, 2] 
        else:
            q_ans = [(place - k) // 2 + 1, 1] 
            
        if abs(row - ans[0]) <= abs(row- q_ans[0]):
            print(*ans)
            return
        else:
            print(*q_ans)
            return
        
        
    elif place + k <= N:
        
        if (place + k) % 2 == 0:
            print((place + k) // 2, 2, sep = " ")
            return 
        
        else:
            print((place + k) // 2 + 1, 1, sep = ' ')
            return
        
    elif place - k > 0:
        
        if (place - k) % 2 == 0:
            print((place - k) // 2, 2, sep = " ")
            return 
        
        else:
            print((place - k) // 2 + 1, 1, sep = ' ')
            return
        
    else:
        print(-1)
        
ans()
    
```

```python
#num 12
s = input()

def ans(s):
    stack = []
    
    for i in s:    
        
        if i == '(' or i == '{' or i == '[':
            stack.append(i)
        
        else:
            
            if stack and ((i == ')' and stack[-1] == '(') or (i == '}' and stack[-1] == '{') or (i == ']' and stack[-1] == '[')):
                stack.pop(-1)
            else:
                stack.append(i)
                
                
                
                
    if stack == []:
        print('yes')
    else:
        print('no')
ans(s)
```

```python
#num 13

s = input()

def ans(s):
    s = s.split()
    stack = []
    
    for i in s:
        
        if stack and not i.isdigit():
            a = int(stack.pop(-1))
            b = int(stack.pop(-1))
            if i == '+':
                stack.append(a + b)    
            elif i == '-':
                stack.append(b - a)
            else:
                stack.append(b * a)
                
        if i.isdigit():
            stack.append(i)
            
            
    print(stack[-1])
ans(s)        
```

```python
# num17

one = [int(i) for i in input().split()]
two = [int(i) for i in input().split()]
count = 0

def ans(one,two,count):
    
    while one and two and count != 10 ** 6:
        one_play_card = one.pop(0)
        two_play_card = two.pop(0)
        

        if (one_play_card == 0 and two_play_card == 9) or (two_play_card == 0 and one_play_card == 9):
            
            if one_play_card == 0 and two_play_card == 9:
                one.append(one_play_card)
                one.append(two_play_card)
            else:
                two.append(one_play_card)
                two.append(two_play_card)

        
        else:
            if one_play_card > two_play_card:
                one.append(one_play_card)
                one.append(two_play_card)
            else:
                two.append(one_play_card)
                two.append(two_play_card)
        
        count += 1
    if count >= 10 ** 6:
        
        print('botva')
        return
    
    if two:
        print('second', count)
        return
    
    print('ferst', count)
    
ans(one,two,count)
    
```

```python
#Num 19
def append_heap(heap,n): # n - индекс элемента
    
    if n == 0:
        return 
    
    if n % 2 == 0:
        
        if heap[(n - 2) // 2] < heap[n]:
            
            heap[(n - 2) // 2], heap[n] = heap[n], heap[(n - 2) // 2]
            append_heap(heap, (n - 2) // 2)
    
    else:
        
        if heap[(n - 1) // 2] < heap[n]:
            
            heap[(n - 1) // 2], heap[n] = heap[n], heap[(n - 1) // 2]
            append_heap(heap, (n - 1) // 2)
        

def reorder_heap(heap, n=0):
    if n * 2 + 1 >= len(heap):
        return
    
    if n * 2 + 2 < len(heap):
        
        if heap[n] < heap[n * 2 + 1] or heap[n] < heap[n * 2 + 2]:

            if heap[n * 2 + 1] >= heap[n * 2 + 2]:
                heap[n * 2 + 1], heap[n] = heap[n], heap[n * 2 + 1]
                reorder_heap(heap, n * 2 + 1)

            else:
                heap[n * 2 + 2], heap[n] = heap[n], heap[n * 2 + 2]
                reorder_heap(heap, n * 2 + 2 )
    else:
        if heap[n] < heap[n * 2 + 1]:
            heap[n * 2 + 1], heap[n] = heap[n], heap[n * 2 + 1]
            reorder_heap(heap, n * 2 + 1)



n = int(input())

heap = []
for _ in range(n):
    l = input().split()

    if len(l) == 2:
        val = int(l[1])

        if heap == []:
            heap.append(val)
        else:
            heap.append(val)
            append_heap(heap,len(heap) - 1)

    else:

        if len(heap) == 1:
            print(heap.pop(-1))
        else:
            heap[0], heap[-1] = heap[-1], heap[0]
            print(heap.pop(-1))  
            reorder_heap(heap, n=0)
                
           
```

```python
#num22
N, k = [int(i) for i in input().split()]

def ans(N,k):
    dp = [1] + [0]*(N - 1)
    
    
    for i in range(1,N):
        for j in range(1,k + 1):
            if i - j < 0:
                break
                
            dp[i] += dp[i-j]   
    return dp[-1]

print(ans(N,k))
    
```

```python
# num24
N = int(input())
stacker = []

for _ in range(N):
    stacker.append([int(i) for i in input().split()])
    


def ans(N, stacker):
   
    dp = [0] * (N + 1) 
    
    for i in range(1,N + 1):
        
        if i == 1:
            dp[i] = stacker[i - 1][0]
            
        elif i == 2:
            dp[i] = min(dp[i-1] + stacker[i - 1][0], dp[i - 2] + stacker[i - 2][1])
            
        else:
            dp[i] = min(dp[i-1] + stacker[i - 1][0], dp[i - 2] + stacker[i - 2][1], dp[i-3] + stacker[i - 3][2])
            
    return dp[-1]
        
            
print(ans(N,stacker))
    
    
```

```python
#num 33
with open('input.txt') as f:
    n,m = [int(i) for i in f.readline().split()]
    graph = {}
    
    node = 1
    fl = True
    for _ in range(m):
        
        node1, node2 = [int(i) for i in f.readline().split()]
        if fl:
            node = node1
            fl = False
        if node1 != node2:
            if node1 not in graph:
                graph[node1] = [node2]
            else:
                graph[node1].append(node2)

            if node2 not in graph:
                graph[node2] = [node1]
            else:
                graph[node2].append(node1)
        else:
            if node1 not in graph:
                graph[node1] = [node2]
            else:
                graph[node1].append(node2)
                
                
    graphs = [0] * n
    Flags = [True]
    
    def dfs(node, color):
        if graphs[node - 1] == 0:
            graphs[node - 1] = 3 - color
            
            if node in graph:
                
                for j in graph[node]:
                    
                    if j == node:
                        continue
                        
                    if graphs[j - 1] == 3 - color:
                        Flags[0] = False
                        
                    else:
                        dfs(j, 3 - color)
    
    for i in range(n):
        if graphs[i] == 0:
            dfs(i+1, 1)
            
    if Flags[0]:
        print('YES')
    else:
        print('NO')
```

```python
#num 35
with open('input.txt') as f:
    n = int(f.readline())
    graph = {}

    for i in range(n):
        for j, num in enumerate(f.readline().split()):
           
            if int(num) and j != i:
                if i not in graph:
                    graph[i] = [j]
                else:
                    graph[i].append(j)
    

    nodes = [0] * n # Вершина и откуда мы в нее пришли
    previos = [0] * n
    ans = []

    def dfs(i,prev):
        if ans:
            return
        
        if nodes[i] == 0:
            nodes[i] = 1
            previos[i] = prev 
            
            if i in graph:
                for j in graph[i]:
                    if prev != j:
                        dfs(j,i)
            
            nodes[i] = 2
        
        elif nodes[i] == 1:
            
            while prev != i:
                
                ans.append(prev + 1)
                prev = previos[prev]
                
            ans.append(prev + 1)
            
    for i in range(n):
        if nodes[i] == 0:
            dfs(i,0)    
            
       
    if ans:
        print('YES')
        print(len(ans))
        print(*ans)
    else:
        print('NO')
                
```

```python
#num 36
with open('input.txt') as f:
    n = int(f.readline())
    graph = {}

    for i in range(n):
        
        for j, num in enumerate(f.readline().split()):
           
            if int(num) and j != i:
                
                if i not in graph:
                    graph[i] = [j]
                
                else:
                    graph[i].append(j)
                    
    L, R = map(int, f.readline().split())
    
    
    stack = [(L - 1, 0)]
    node_set = set()
    ans = 0
    while stack:
        
        
        node, count = stack.pop(0)
        
        if node == R - 1:
            ans = count
            break
            
        if node in graph:
            
            if node not in node_set:
                node_set.add(node)
                for i in graph[node]:
                    stack.append((i, count + 1))

    def answer(ans):
        if L == R:
            print(0)
            return
        
        if ans:
            print(ans)
        else:
            print(-1)
    
                
    answer(ans)
```

```python
#num 37
with open('input.txt') as f:
    n = int(f.readline())
    graph = {}

    for i in range(n):
        
        for j, num in enumerate(f.readline().split()):
           
            if int(num) and j != i:
                
                if i not in graph:
                    graph[i] = [j]
                
                else:
                    graph[i].append(j)
                    
    L, R = map(int, f.readline().split())
    
    
    stack = [(L - 1, 0)]
    node_set = set()
    ans = []
    previos = [0] * n
    counts = [float('inf')] * n
    while stack:
        
        
        node, count = stack.pop(0)
        
        if node == R - 1:
            while count != -1:
                
                ans.append(node + 1)
                node = previos[node]
                count -= 1
            break
            
        if node in graph:
            
            if node not in node_set:
                node_set.add(node)
                for i in graph[node]:
                    
                    if counts[i] > count:
                        previos[i] = node
                        counts[i] = count
                        
                    stack.append((i, count + 1))
    
    def answer(ans):
        if L == R:
            print(0)
            return
        
        if ans:
            print(len(ans) - 1)
            print(*ans[::-1])
        else:
            print(-1)
    
                
    answer(ans)
```

```python
#num 40
with open('input.txt') as f:
    N = int(f.readline())
    M = int(f.readline())
  
    graph = {}
    numlines = {}
    line = 1
    count_stations = 0
    for _ in range(M):
        
        lines = f.readline().split()
        
        num_station = int(lines[0])
        stations = [int(i) for i in lines[1:]]
        
        for i in range(num_station):
            
            if stations[i] not in numlines:
                numlines[stations[i]] = set([line])
            else:
                numlines[stations[i]].add(line)
            
            if i == 0:
                continue
            
            if stations[i] not in graph: 
                graph[stations[i]] = [stations[i - 1]]
            else:
                if stations[i-1] not in graph[stations[i]]:
                    graph[stations[i]].append(stations[i-1])
                
            if stations[i-1] not in graph:
                graph[stations[i-1]] = [stations[i]]
            else:
                if stations[i] not in graph[stations[i - 1]]:
                    graph[stations[i - 1]].append(stations[i])
        line += 1
        
        
    
    
    A,B = map(int, f.readline().split())
    
    stack = [(A, 0, numlines[A])] # Пересечение множест не нулевое - станция на той же линии
    ans = float('inf')
    
    nodes = [0] * N
    
    
    while stack:

        station, count, num_lines = stack.pop(0)  # numlines - то откуда пришли
        
        if nodes[station - 1] == 0:
            nodes[station - 1] = 1
        
       
        if station == B:
            ans = min(count, ans)
            continue


        for i in graph[station]:


            if num_lines & numlines[i] == set():
                if nodes[i-1] == 0:
                    stack.append((i, count + 1, numlines[station] & numlines[i]))

            else:
                if nodes[i-1] == 0:
                    stack.append((i, count, num_lines))

          
        
    
    if ans == float('inf'):
        print(-1)
    else:
        print(ans)
                
    
    
        
        
```
